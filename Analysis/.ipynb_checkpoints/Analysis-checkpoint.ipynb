# Data preparation (sample)

```python
# Define a function that reads and appends chunks:
def read_csv_chunk(csv_file, var_list):
    data_iterator = pd.read_csv(csv_file, chunksize=100000, usecols=var_list)
    
    chunk_list = []
    # Each chunk is in dataframe format
    for data_chunk in data_iterator:  
        consolidated_chunk = data_chunk.groupby(var_list).size()
        chunk_list.append(consolidated_chunk)
    return pd.concat(chunk_list)
```

```python
# Oct and Nov user lists
Oct_user = read_csv_chunk("../Data/2019-Oct.csv", ['user_id']).index.unique()
Nov_user = read_csv_chunk("../Data/2019-Nov.csv", ['user_id']).index.unique()

# User retention rate from Oct to Nov
print('Oct total user: {}; Churned user: {}; Retention rate from Oct to Nov: {}%' \
        .format(len(Oct_user), len(Oct_user.difference(Nov_user)), 100 - 100*len(Oct_user.difference(Nov_user))/len(Oct_user)))
```

```python
# Sample 1% of Oct users for processing
perc = 0.01
select_num = random.sample(range(len(Oct_user)), math.ceil(len(Oct_user)*perc))
Oct_user_selected = Oct_user[select_num]
```

# Feature construction

```python
df_oct_sample = pd.read_csv("../Data/2019-Oct-sample.csv", index_col=0)
```

```python
df_oct_sample['event_date'] = pd.to_datetime(df_oct_sample['event_time'].str[:10])
```

```python
# Collapse to user level
df = df_oct_sample[['user_id', 'f_retention', 'user_session']].drop_duplicates() \
    .groupby(['user_id', 'f_retention']).agg(num_session = ('user_session', 'count'))
```

```python
# Number of each type of events
num_events_per_user = df_oct_sample.groupby(['user_id', 'f_retention', 'event_type']).size().unstack('event_type').fillna(0)
col_names =['num_' + col for col in num_events_per_user.columns]
df[col_names] = num_events_per_user
```

```python
# Value of purchase, if any
df_oct_sample['purchase_amount'] = np.where(df_oct_sample.event_type == 'purchase', df_oct_sample['price'], np.nan)
df['purchase_amount'] = df_oct_sample.groupby(['user_id', 'f_retention'])['purchase_amount'].sum()
```

```python
# Date of month of user's first activity
df['first_event_day_of_month'] = df_oct_sample.assign(dom = df_oct_sample.event_date.dt.day).groupby(['user_id', 'f_retention']).dom.min()
```

```python
# Number of unique products viewed, added to cart, or purchased
num_products_per_user = df_oct_sample[['user_id', 'f_retention', 'event_type', 'product_id']].drop_duplicates() \
                            .groupby(['user_id', 'f_retention', 'event_type']).size().unstack('event_type').fillna(0)
col_names =['num_product_' + col for col in num_events_per_user.columns]
df[col_names] = num_products_per_user
```

```python
# Number of unique categories viewed, added to cart, or purchased
num_cats_per_user = df_oct_sample[['user_id', 'f_retention', 'event_type', 'category_id']].drop_duplicates() \
                            .groupby(['user_id', 'f_retention', 'event_type']).size().unstack('event_type').fillna(0)
col_names =['num_category_' + col for col in num_events_per_user.columns]
df[col_names] = num_cats_per_user
```

```python
# Avg price of products each user viewed
df['avg_price_view'] = df_oct_sample.groupby(['user_id', 'f_retention', 'event_type'])['price'].mean().unstack('event_type')[['view', 'cart', 'purchase']].bfill(axis=1).iloc[:, 0]
```

```python
# Encode category by its number of times being viewed (popularity)
category_id_encoded = df_oct_sample[df_oct_sample.event_type == 'view'].groupby('category_id').size()
df_oct_sample['category_id_encoded'] = df_oct_sample['category_id'].map(category_id_encoded)
```

```python
# Encode category_id and consolidate to user level
# Method 1: for each category and user, count the number of events associated (i.e., view, cart, purchase)
user_cat_level_cnt = df_oct_sample.groupby(['user_id', 'f_retention', 'category_id']).size().unstack('category_id').fillna(0)
col_names =['cat_id_cnts' + str(col) for col in user_cat_level_cnt.columns]
df[col_names] = user_cat_level_cnt
```

# Logit model (sample)

```python
def eval_report(x, y):
    print("Accuracy score {:.2f} %\n".format(model.score(x,y)*100))
    plot_confusion_matrix(model,x,y,cmap=plt.cm.Blues)
    print(classification_report(y,model.predict(x)))
```

```python
from sklearn.pipeline import Pipeline

# Build Model here
# model = LogisticRegression(solver='liblinear', penalty='l1')
# model.fit(x_train, y_train)
# Build Model here
model = Pipeline(steps=[('minmax', MinMaxScaler()),('lr', LogisticRegression(penalty='l2'))])
model.fit(x_train, y_train)
```

### Build a learning curve

```python
train_sizes = range(2, len(x_train), 1000)
```

```python
ls_train=[]
ls_test=[]
for ts in train_sizes:
    sample = random.sample(range(len(x_train)), ts)
    model = LogisticRegression(solver='liblinear', penalty='l1')
    model.fit(x_train[sample], y_train.iloc[sample])
    ls_train.append(1- accuracy_score(y_train.iloc[sample], model.predict(x_train[sample])))
    ls_test.append(1-accuracy_score(y_test, model.predict(x_test)))
```

```python
plt.plot(train_sizes, ls_train, label = 'Training error')
plt.plot(train_sizes, ls_test, label = 'Test error')
plt.legend()
```
